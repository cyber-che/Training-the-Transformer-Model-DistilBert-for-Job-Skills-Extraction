```python
import pandas as pd
from collections import Counter


# x_file = './tech_industry_top500skills_substituted.csv'
x_file="iteration4_tech_job.csv"
df = pd.read_csv(x_file)


if 'filtered_skills' in df.columns:
    # Split the comma-separated skills and flatten the list of words
    all_skills = df['filtered_skills'].str.split(',').explode().str.strip()

    # Count frequency of each word
    word_counts = Counter(all_skills)

    word_freq_df = pd.DataFrame(word_counts.items(), columns=['Skill', 'Frequency'])

    # Save results to CSV
    output_file = 'word_frequencies_synonym.csv' 
    word_freq_df.to_csv(output_file, index=False)

    print(f"Word frequencies have been written to {output_file}")
else:
    print("The 'filtered_skills' column does not exist in the CSV file.")
```
